```python
import requests as req
from bs4 import BeautifulSoup as bs
from pprint import pprint
```

```python
url="https://www.ntu.edu.tw/"

header ={
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'
}

res=req.get(url,headers = header)

list_news = []



soup=bs(res.text,'lxml')
new = soup.select_one('#news')


for news in new.select('a'):
    list_news.append(news['href'])
    
    
    
del list_news[0:1]
print(list_news)

for index,link in enumerate(list_news):
    res=req.get(link)
    res.encoding='utf-8'
    link_soup=bs(res.text,'lxml')
   
    inner=link_soup.select_one('div.col-md-8.welcom-grid-text').get_text()
    print(inner)
    
    
```
